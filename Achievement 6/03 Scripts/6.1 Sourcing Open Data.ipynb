```python
#checking for null values
citibike.isnull().sum()
```

```python
#Creating a subset data of nan values in birth_year column
NaN=citibike[citibike['birth_year'].isnull() ==True]
```

```python
#Checking NaN values shape
NaN.shape
```

```python
#Dropping NaN values in birth_year Column
citibike= citibike[citibike['birth_year'].isnull()== False]
```

```python
#Checking if the NaN values are dropped
citibike.shape
```

```python
#Changing data type
citibike['birth_year']=citibike['birth_year'].astype('int64')
```

```python
#Convert date columns to datetime
citibike['start_time']=pd.to_datetime(citibike['start_time'])
citibike['end_time']=pd.to_datetime(citibike['end_time'])
```

```python
citibike.dtypes
```

```python
#Checking the changes made
citibike.head()
```

```python
#Creating a dictionnary to map the values to gender labels
gender_mapping={0:'Unkown',1:'Male',2:'Female'}
```

```python
#Replacing the numerical values with gender labels
citibike['gender']=citibike['gender'].replace(gender_mapping)
```

```python
#Confirming the changes made
citibike.head()
```

```python
#Renaming column names
citibike.rename(columns={'weekday':'day_of_the_week','start_time':'ride_start_time','end_time':'ride_end_time'},inplace=True)
```

```python
#Checking changes made
citibike.head()
```

```python
#Dropping columns
columns_to_drop=['trip_id','bike_id']
citibike.drop(columns=columns_to_drop,inplace=True)
```

```python
#Checking the changes made
citibike.head()
```
