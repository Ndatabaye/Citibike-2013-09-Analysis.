# 2. Picking Variables

```python
#Picking Variables to use and creating a subset
citibike1=citibike_clean[['start_hour','start_station_latitude','start_station_longitude','end_station_latitude','end_station_longitude','trip_duration','gender']]
```

# 3. Creating a colored Correlation Heatmap

```python
#Checking for correlation
citibike1.corr()
```

```python
#Defining correlation matrix
correlation_matrix = citibike1.corr()
# Creating a subplot with Matplotlib
f, ax = plt.subplots(figsize=(7,7))
# Creating the correlation heatmap using Matplotlib's matshow
heatmap = ax.matshow(correlation_matrix, cmap='coolwarm')
# Adding annotations with correlation values
for i in range(correlation_matrix.shape[0]):
    for j in range(correlation_matrix.shape[1]):
        ax.text(x=j, y=i, s=f"{correlation_matrix.iloc[i, j]:.2f}",
                ha='center', va='center', fontsize=12, color='black')
# Customizing the colorbar
cbar = plt.colorbar(heatmap)
cbar.ax.tick_params(labelsize=12)
#Defining path
path=r'/Users/guillaumemac/Achievement 6/04 Analysis/Visualizations/correlation_heatmap.png'
# Set axis labels and title
ax.set_xticks(range(len(correlation_matrix.columns)))
ax.set_yticks(range(len(correlation_matrix.columns)))
ax.set_xticklabels(correlation_matrix.columns, fontsize=12, rotation=90)
ax.set_yticklabels(correlation_matrix.columns, fontsize=12)
ax.set_title('Correlation Matrix Heatmap', fontsize=14)
#Saving the visualization
plt.savefig(path)
```

### There seems to not be any particular strong correlation between the variables.The strongest correlation observed is between start_station latitude and end_station latitude; start_station_longitude and end_station_longitude;end_station_latitude and start_station_latitude;end_station_longitude and start_station_longitude.
